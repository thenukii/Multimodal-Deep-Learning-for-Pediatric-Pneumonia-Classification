```python
base_dir = '/content/drive/MyDrive/IIT/FYP/Implementation/data/Chest_Xray/chest_xray_split'
normal_dir = f'{base_dir}/train/NORMAL'
augmented_dir = f'{base_dir}/train_augmented/NORMAL'
```

```python
import os
os.makedirs(augmented_dir, exist_ok=True)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.1,
    height_shift_range=0.1,
    shear_range=0.1,
    zoom_range=0.2,
    horizontal_flip=True
)
```

```python
i = 0
target_count = 2200  # number of extra images to balance dataset

for batch in datagen.flow_from_directory(
    base_dir + '/train/',
    classes=['NORMAL'],
    target_size=(224,224),
    batch_size=16,
    save_to_dir=augmented_dir,
    save_prefix='aug',
    save_format='jpg'):

    i += 16
    if i >= target_count:
        break
```

```python
import os

normal_original = len(os.listdir(normal_dir))
normal_augmented = len(os.listdir(augmented_dir))

print("Original Normal images:", normal_original)
print("New Augmented images:", normal_augmented)
print("Total Normal images:", normal_original + normal_augmented)
```

```python
import shutil

normal_train_dir = "/content/drive/MyDrive/IIT/FYP/Implementation/data/Chest_Xray/chest_xray_split/train/NORMAL"

for file in os.listdir(augmented_dir):
    src = os.path.join(augmented_dir, file)
    dst = os.path.join(normal_train_dir, file)
    shutil.copy(src, dst)

print("Augmented images merged successfully!")
```
